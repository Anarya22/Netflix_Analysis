**Q.8 In how many movie/shows, Tom Cruise was cast?**

```python
df.head(2)
```

```python
df[df['Cast']=='Tom Cruise']
```

```python
# df[df['Cast'].str.contains('Tom Cruise')]  ----> IT WILL THROW AN ERROR BECAUSE OF NAN VALUES
```

```python
# Impute NaN values in the 'Cast' column with a relevant string
df['Cast'] = df['Cast'].fillna('Unknown')
df[df['Cast'].str.contains('Tom Cruise')]
```

```python
df.isnull().sum()
```

```python
colm_2_impute = ['Director','Country','Release_Date','Rating','DateN','Year']
df_new = df.copy()
# df_new[colm_2_impute] = df_new[colm_2_impute].fillna('Unknown')

for col in colm_2_impute:
    if df_new[col].dtype == 'object':  # If the column is of type 'object' (string)
        df_new[col] = df_new[col].fillna('Unknown')
    else:  # If the column is of a numeric type
        df_new[col] = df_new[col].fillna(0)
df_new
```

```python
df_new.isnull().sum()
```
